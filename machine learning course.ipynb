```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel(r"C:\Users\User\Desktop\Linear Regressio.xlsx")
```

```python
df
```

```python
w = df['Weight']
p = df['Price']
```

```python
w_mean = np.mean(w)
```

```python
p_mean = np.mean(p)
```

```python
x = w - w_mean
y = p - p_mean
```

```python
m = np.sum(x * y)/np.sum(x**2)
```

```python
print(m)
```

```python
c = p_mean - m*w_mean
```

```python
print(c)
```

```python
predict_6 = m*6+ c
```

```python
print(predict_6)
```

```python
plt.scatter(w,p)
plt.show()
```
